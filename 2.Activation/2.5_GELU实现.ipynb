### 1.函数实现

```python
import numpy as np
from scipy.special import erf
def gelu(x):
    """Compute the GELU activation function value for a given input x.
    
    The GELU function is defined as:
        gelu(x) = 0.5 * x * (1 + tanh(sqrt(2/pi) * (x + 0.044715 * x^3)))
    """
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))
```

### 2.函数图像

```python
# 画图GeLU函数
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = [gelu(i) for i in x]
plt.plot(x, y)
plt.title("GELU Function")
plt.xlabel("x")
plt.ylabel("GELU(x)")
plt.grid()
plt.show()
```
