### 1.函数实现

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python

def swish(x):
    # Swish(x) = x * sigmoid(x)
    return x * sigmoid(x)
```

### 2.函数图像

```python
# 画图swish函数
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = [swish(i) for i in x]
plt.plot(x, y)
plt.title("Swish Function")
plt.xlabel("x")
plt.ylabel("Swish(x)")
plt.grid()
plt.show()
```
