### 1.函数实现

```python
import numpy as np
def relu(x):
    return np.maximum(0, x)
```

### 2.函数图像

```python
# 画图sigmond函数
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = [relu(i) for i in x]
plt.plot(x, y)
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.grid()
plt.show()
```
