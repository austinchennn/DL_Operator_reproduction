### 1.函数实现

```python
import math


def softmax(x: list[list[float]]) -> list:
    """

    x: 输入数据 list
    """

    y = []
    for row in x:
        # 稳定性找最大的那个数
        row_max = max(row)
        # 计算这行每个数的e^val
        exp_row = [math.exp(val - row_max) for val in row]
        # 计算分母
        sum_exp = sum(exp_row)
        # 归一化：每个元素除以总和
        probs_row = [val / sum_exp for val in exp_row]
        y.append(probs_row)
    return y

```

```python
# 画图：展示两个输入的softmax
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 200)

# 稳定版本：输入 [0, x]
y_stable = []
for val in x:
    result = softmax([[0.0, val]])
    y_stable.append(result[0][1])


plt.figure(figsize=(10, 6))
plt.plot(x, y_stable, label='Stable Softmax', linewidth=2.5, color='blue')
plt.xlabel('Input [0, x]', fontsize=12)
plt.ylabel('P(x) - Probability', fontsize=12)
plt.title('Softmax: Stable vs Unstable', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
