### 1.函数实现

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
import numpy as np
def sinlu(x, a=1.0, b=1.0):
    """
    x: 输入向量
    a: 振幅 (Amplitude)，可学习参数
    b: 频率 (Frequency)，可学习参数
    """
    # 公式: (x + a * sin(bx)) * sigmoid(x)
    return (x + a * np.sin(b * x)) * sigmoid(x)
```

### 2.函数图像

```python
# 画图SinLU函数
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = [sinlu(i) for i in x]
plt.plot(x, y)
plt.title("SinLU Function")
plt.xlabel("x")
plt.ylabel("SinLU(x)")
plt.grid()
plt.show()
```
