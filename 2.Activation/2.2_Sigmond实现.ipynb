### 1.函数实现
##### Sigmoid 实际上就是 Softmax 在二分类情况下的特例

```python
def sigmond(x):
    """Compute the Sigmond function value for a given input x.

    The Sigmond function is defined as:
        sigmond(x) = 1 / (1 + exp(-x))
    """
    import numpy as np
    return 1 / (1 + np.exp(-x))
```

### 2.函数图像

```python
# 画图sigmond函数
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = [sigmond(i) for i in x]
plt.plot(x, y)
plt.title("Sigmond Function")
plt.xlabel("x")
plt.ylabel("sigmond(x)")
plt.grid()
plt.show()
```
